## This is an example of how to create Frequency Tables for continuous quantitative data, after a previous categorization, using Python, Pandas, NumPy and Matplotlib

#### * the data from the .csv file can be of 50 rows, 10000 rows or 1 million rows, that's no problem for NumPy and Pandas

```python
# importing libs and setting default plot style
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("dark_background")
import pandas as pd
import numpy as np
```

```python
# importing data into a Pandas Dataframe
income_dataframe = pd.read_csv("https://raw.githubusercontent.com/danielpm1982/sample-jupyter-notebook-statistics/master/income.csv")
# showing top 5 rows
income_dataframe.head()
```

```python
print(f"{len(income_dataframe)} rows")
```

```python
# reformatting INCOME (US$) data into int and without thousands commas (for plotting on a graph)
income_dataframe["INCOME (US$)"] = income_dataframe["INCOME (US$)"].replace(',', '', regex=True)
income_dataframe["INCOME (US$)"] = income_dataframe["INCOME (US$)"].astype("float").astype("int")
# showing top 10 rows
income_dataframe.loc[:10]
```

```python
# plotting a matplotlib graph representing the Client's Income data in US$ versus the respective Client ID

# setting the marker (dot) size according to n:
if(len(income_dataframe)<=100):
    markersize=50
elif(len(income_dataframe)>100 and len(income_dataframe)<=1000):
    markersize=10
else:
    markersize=1

# creating the Scatter graph
income_dataframe.plot(kind="scatter", x="CLIENT_ID", y="INCOME (US$)", s=markersize, c=income_dataframe["INCOME (US$)"], 
                      cmap="gist_rainbow", alpha=0.8, title=f"Income Scatter Graph (n={len(income_dataframe)})", 
                      xlabel="CLIENT_ID", ylabel="INCOME (US$)", grid=True, xlim=[0,len(income_dataframe)])
plt.show()
```

```python
# setting columns for the new Dataframe - the Frequency Table Dataframe
frequency_table_columns=["INCOME_INTERVAL", "ABSOLUTE_FREQUENCY", "RELATIVE_FREQUENCY", "CUMULATIVE_ABSOLUTE_FREQUENCY", "CUMULATIVE_RELATIVE_FREQUENCY"]
frequency_table_columns
```

```python
# creating Dataframe instance with column names and empty data
frequency_table_dataframe = pd.DataFrame(columns=frequency_table_columns)
frequency_table_dataframe
```

```python
# setting up and populating INCOME_INTERVAL column at frequency_table_dataframe - categorization (quantitative to qualitative data)
# INCOME_INTERVAL here consists in the derived category intervals, from the continuous numerical income data, so that
# frequencies can then be calculated for each one. This is called categorization
# frequency tables can't be set based on continuous numerical data, they must be based on discrete numerical or on qualitative data sets
# the code blocl below simply sets those intervals and resets all other column cells to NaN for avoiding showing cached values
frequency_table_dataframe["INCOME_INTERVAL"] = ["0-2000","2001-4000","4001-6000","6001-8000","8001-10000", "10001-12000", "12001-15000"]
frequency_table_dataframe["ABSOLUTE_FREQUENCY"] = np.nan
frequency_table_dataframe["RELATIVE_FREQUENCY"] = np.nan
frequency_table_dataframe["CUMULATIVE_ABSOLUTE_FREQUENCY"] = np.nan
frequency_table_dataframe["CUMULATIVE_RELATIVE_FREQUENCY"] = np.nan
frequency_table_dataframe
```

```python
# setting and populating ABSOLUTE_FREQUENCY values at frequency_table_dataframe
# creating a function for returning a Series with all income data between lower and upper limits, from the income_dataframe table
def get_income_interval_Series(lower_limit_inclusive, upper_limit_inclusive):
    return income_dataframe[(income_dataframe["INCOME (US$)"]>=lower_limit_inclusive) & (income_dataframe["INCOME (US$)"]<=upper_limit_inclusive)]
# creating a function for returning the number of results returned for each lower-upper income interval above (absolute frequency)
def get_income_interval_Series_absolute_frequency(lower_limit_inclusive, upper_limit_inclusive):
    return len(get_income_interval_Series(lower_limit_inclusive, upper_limit_inclusive))
# populating the ABSOLUTE_FREQUENCY column with the values calculated above for each interval category and resetting remaining 
# values as NaN for avoiding showing cached values
frequency_table_dataframe["ABSOLUTE_FREQUENCY"] = [get_income_interval_Series_absolute_frequency(0,2000), get_income_interval_Series_absolute_frequency(2001,4000), get_income_interval_Series_absolute_frequency(4001,6000), get_income_interval_Series_absolute_frequency(6001,8000), get_income_interval_Series_absolute_frequency(8001,10000), get_income_interval_Series_absolute_frequency(10001,12000), get_income_interval_Series_absolute_frequency(12001,15000)]
frequency_table_dataframe["RELATIVE_FREQUENCY"] = np.nan
frequency_table_dataframe["CUMULATIVE_ABSOLUTE_FREQUENCY"] = np.nan
frequency_table_dataframe["CUMULATIVE_RELATIVE_FREQUENCY"] = np.nan
frequency_table_dataframe
```

```python
# setting and populating RELATIVE_FREQUENCY values at frequency_table_dataframe, by dividing the total number of occurencies of data,
# from each income interval, by the total number of income values, at the income_dataframe table (n)
# resetting remaining values as NaN for avoiding showing cached values
def get_income_interval_Series_relative_frequency(lower_limit_inclusive, upper_limit_inclusive):
    return get_income_interval_Series_absolute_frequency(lower_limit_inclusive, upper_limit_inclusive)/len(income_dataframe["INCOME (US$)"])
frequency_table_dataframe["RELATIVE_FREQUENCY"]=[get_income_interval_Series_relative_frequency(0,2000), get_income_interval_Series_relative_frequency(2001,4000), get_income_interval_Series_relative_frequency(4001,6000), get_income_interval_Series_relative_frequency(6001,8000), get_income_interval_Series_relative_frequency(8001,10000), get_income_interval_Series_relative_frequency(10001,12000), get_income_interval_Series_relative_frequency(12001,15000)]
frequency_table_dataframe["CUMULATIVE_ABSOLUTE_FREQUENCY"] = np.nan
frequency_table_dataframe["CUMULATIVE_RELATIVE_FREQUENCY"] = np.nan
frequency_table_dataframe
```

```python
# setting and populating CUMULATIVE_ABSOLUTE_FREQUENCY values at frequency_table_dataframe
# resetting remaining values as NaN for avoiding showing cached values
def get_income_interval_Series_cumulative_absolute_frequency(lower_limit_inclusive, upper_limit_inclusive):
    return get_income_interval_Series_absolute_frequency(0, upper_limit_inclusive)
frequency_table_dataframe["CUMULATIVE_ABSOLUTE_FREQUENCY"] = [get_income_interval_Series_cumulative_absolute_frequency(0,2000), get_income_interval_Series_cumulative_absolute_frequency(2001,4000), get_income_interval_Series_cumulative_absolute_frequency(4001,6000), get_income_interval_Series_cumulative_absolute_frequency(6001,8000), get_income_interval_Series_cumulative_absolute_frequency(8001,10000), get_income_interval_Series_cumulative_absolute_frequency(10001,12000), get_income_interval_Series_cumulative_absolute_frequency(12001,15000)]
frequency_table_dataframe["CUMULATIVE_RELATIVE_FREQUENCY"] = np.nan
frequency_table_dataframe
```

```python
# setting and populating CUMULATIVE_RELATIVE_FREQUENCY values at frequency_table_dataframe
def get_income_interval_Series_cumulative_relative_frequency(lower_limit_inclusive, upper_limit_inclusive):
    return get_income_interval_Series_relative_frequency(0, upper_limit_inclusive)
frequency_table_dataframe["CUMULATIVE_RELATIVE_FREQUENCY"] = [get_income_interval_Series_cumulative_relative_frequency(0,2000), get_income_interval_Series_cumulative_relative_frequency(2001,4000), get_income_interval_Series_cumulative_relative_frequency(4001,6000), get_income_interval_Series_cumulative_relative_frequency(6001,8000), get_income_interval_Series_cumulative_relative_frequency(8001,10000), get_income_interval_Series_cumulative_relative_frequency(10001,12000), get_income_interval_Series_cumulative_relative_frequency(12001,15000)]
frequency_table_dataframe
```
